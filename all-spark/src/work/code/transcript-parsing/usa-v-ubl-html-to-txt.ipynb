```python
import urllib
import glob
import os
from bs4 import BeautifulSoup
```

```python
for filePath in glob.glob("/transcripts/usa-vs-ubl/*.htm"):
    with open(filePath, 'r') as file:
        fileName = os.path.basename(file.name)
        fileNameWithoutExt = os.path.splitext(fileName)[0]
        #print(fileNameWithoutExt)
        content = file.read()
        soup = BeautifulSoup(content, 'html.parser')
        try:
            transcript = soup.body.pre.text
            with open("/transcripts/usa-vs-ubl/"+fileNameWithoutExt+".txt",'w') as outputFile:
                outputFile.write(transcript)        
        except:
            print("Oops! " + fileName)
            raise        
    
    #^ {63,}\d+$
```
